```python
import pandas as pd
import os
```

```python
# Get all the files downloaded and stored in WHRData
files=os.listdir('WHRData')
```

```python
#Create a list of files with Data in the name
files=[f for f in files if 'Data' in f]
files
```

```python
#Print Years from file names
for file in files:
    print(file.split('.')[0].replace('WHR','').replace('Data','').strip())
```

```python
DFlist=[] # Emptylist
for file in files: #for all files in the list
    print(f'Processing {file}')
    xls1=pd.ExcelFile("WHRData\\" + file) #Read excel file
    if '2015' in file:
        df1 = pd.read_excel(xls1,'Data for Figure2.2',skiprows=1)
        df1 = df1[['country','Ladder score','Logged GDP per capita','Social support','Healthy life expectancy',
        'Freedom to make life choices','Generosity','Perceptions of corruption']]
        df1.rename(columns=({'Ladder score':'Score','country':'Country'}),inplace=True)
    else:
        if ('2016' in file) or ('2018' in file):
            df1 = pd.read_excel(xls1,'Figure2.2')
        if '2017' in file:
            df1 = pd.read_excel(xls1,'Figure2.2 WHR 2017')
        if '2019' in file:
            df1 = pd.read_excel(xls1,'Figure2.6')
        if ('2020' in file) or ('2021' in file):
            df1 = pd.read_excel(xls1,'Sheet1')
            df1.rename(columns=({'Ladder score':'Score','Country name':'Country'}),inplace=True)
        file_year=file.split('.')[0].replace('WHR','').replace('Data','').strip()
        #For files in years 2016-19
        if (file_year=='2016' or file_year=='2017' or file_year=='2018' or file_year=='2019'):
            df1 = df1[['Country', 'Happiness score','Explained by: GDP per capita', 'Explained by: Social support',
           'Explained by: Healthy life expectancy','Explained by: Freedom to make life choices',
           'Explained by: Generosity', 'Explained by: Perceptions of corruption']]
            df1.rename(columns=({'Happiness score':'Score','Explained by: GDP per capita':'Logged GDP per capita',
                            'Explained by: Social support':'Social support',
                            'Explained by: Healthy life expectancy':'Healthy life expectancy',
                            'Explained by: Freedom to make life choices':'Freedom to make life choices',
                            'Explained by: Generosity':'Generosity',
                            'Explained by: Perceptions of corruption':'Perceptions of corruption'
                           }),inplace=True)
    df1.reset_index(inplace=True)
    df1.rename(columns=({'index':'Rank'}),inplace=True)
    df1['Rank']=df1['Rank']+1
    df1['Year']=file.split('.')[0].replace('WHR','').replace('Data','').strip()
    df1=df1[['Country','Score','Rank','Year','Logged GDP per capita','Social support','Healthy life expectancy',
        'Freedom to make life choices','Generosity','Perceptions of corruption']]
    DFlist.append(df1)
    #break
```
