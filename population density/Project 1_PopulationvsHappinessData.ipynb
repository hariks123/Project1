# "How Does Population Density Compare to Happiness?"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
import warnings
warnings.filterwarnings("ignore")

# read csv files
population_df = pd.read_csv("API_EN.POP.DNST_DS2_en_csv_v2_3052789.csv")
happ_pop21 = pd.read_csv("./csvData (3).csv")



```

# Data Gathering, Merging and Cleanup
<b>Data and Modeling Approach<b><br />
    
<i>Data gathered from various websites:https://data.worldbank.org/indicator/EN.POP.DNST, https://worldpopulationreview.com/country-rankings/happiness-by-country, https://worldhappiness.report/archive/<i>
    

```python
# sorting data
sorted_pop = population_df.sort_values(by = "Country Name", axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last')

sorted_pop
```

```python
sorted_happ = happ_pop21.sort_values(by = "country", axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last')
sorted_happ
```

```python
# rename column for merging 
sorted_pop.rename(columns={'Country Name': 'country'}, inplace=True)
sorted_pop.head()
```

```python
# merge two population data sets to add 2021
merged_popdata = sorted_pop.merge(sorted_happ, on='country',how='left')
merged_popdata
```

```python
# Create new empty dataframe for population data
populationsDf = pd.DataFrame(columns = ["Country", "Year", "Population density (people per sq. km of land area)"])

# Append a new row to the new dataframe for every row/year combination in originaal dataframe
for index, row in merged_popdata.iterrows():
    for year in range(2015, 2021):
        if str(year) in merged_popdata.columns:
            newrow = [row["country"], year, row[str(year)]]
            populationsDf.loc[len(populationsDf)] = newrow

populationsDf
```

```python
happData = pd.read_csv("WHR2015-21.csv")
happData
```

```python
happPop_data = populationsDf.merge(happData, on=['Country','Year'],how='left')
happPop_data.head()
```

```python
# drop Nan values
popHapp_data = happPop_data.dropna()
```

```python
# write data to csv file
popHapp_data.to_csv("HappinessandPopulation.csv")
```

# Data Visualization and Analysis
<b>Time to plot!
<i>Here is where the data is plotted to show correlation between happiness and population density

```python
# showing population density in relation to happiness score


fig, ax = plt.subplots()


plt.xlim(0, 700)
plt.ylim(0, 8)

(m, b, r, p, stderr) = linregress(popHapp_data['Population density (people per sq. km of land area)'], popHapp_data["Score"])
y_regress = m * popHapp_data['Population density (people per sq. km of land area)'] + b

# Plot line of regression
plt.plot(popHapp_data['Population density (people per sq. km of land area)'], y_regress, color="red", label = "line fit for all years")
plt.annotate(f"y = {m:.2f}x + {b:.2f}", color="red", xy=(65, 7.5))
plt.annotate(f"r = {r:.2f}", color="red", xy=(65, 7.0))

# create scatter plot

b = plt.scatter(popHapp_data['Population density (people per sq. km of land area)'], popHapp_data["Score"],c=popHapp_data['Year'])
legend = plt.legend(*b.legend_elements(), title ='Year')
ax.add_artist(legend)
ax.legend(bbox_to_anchor=(1.1, 1.05))
plt.legend([b],labels=['line fit for all years'])

plt.title("Happiness vs Population Density")
plt.xlabel('Population density (people per sq. km of land area)')
plt.ylabel("Happiness Score")
plt.savefig("populationandhappinessscore.png")
plt.show()
```

```python
# countries with happiness score higher than 6

highest_pop_df = popHapp_data.loc[popHapp_data["Score"] > 6, [
    ("Population density (people per sq. km of land area)"), "Country", "Year", "Score"]]

highest_pop_df.sort_values(['Population density (people per sq. km of land area)', 'Country'], ascending=False).groupby('Year').head()

highest_pop_df.to_csv("HighestScoredHappiness.csv")
```

```python

fig, ax = plt.subplots()


plt.xlim(0, 700)
plt.ylim(0, 8)

(m, b, r, p, stderr) = linregress(highest_pop_df['Population density (people per sq. km of land area)'], highest_pop_df["Score"])
y_regress = m * highest_pop_df['Population density (people per sq. km of land area)'] + b

# Plot line of regression
plt.plot(highest_pop_df['Population density (people per sq. km of land area)'], y_regress, color="red", label = "line fit for all years")
plt.annotate(f"y = {m:.2f}x + {b:.2f}", color="red", xy=(65, 7.5))
plt.annotate(f"r = {r:.2f}", color="red", xy=(65, 7.0))

# create scatter plot

b = plt.scatter(highest_pop_df['Population density (people per sq. km of land area)'], highest_pop_df["Score"],c=highest_pop_df['Year'])
legend = plt.legend(*b.legend_elements(), title ='Year')
ax.add_artist(legend)
ax.legend(bbox_to_anchor=(1.1, 1.05))
plt.legend([b],labels=['line fit for all years'])

plt.title("Population Density for Highest Happiness Scores")
plt.xlabel('Population density (people per sq. km of land area)')
plt.ylabel("Happiness Score")
plt.savefig("highestpopulationandhappinessscore.png")
plt.show()
```

```python
# countries with happiness score lower than 4

lower_pop_df = popHapp_data.loc[popHapp_data["Score"] < 4, [
    ("Population density (people per sq. km of land area)"), "Country", "Year", "Score"]]

lower_pop_df.sort_values(['Population density (people per sq. km of land area)', 'Country'], ascending=False).groupby('Year').head()

lower_pop_df.to_csv("LowestestScoredHappiness.csv")
```

```python

fig, ax = plt.subplots()


plt.xlim(0, 700)
plt.ylim(0, 8)

(m, b, r, p, stderr) = linregress(lower_pop_df['Population density (people per sq. km of land area)'], lower_pop_df["Score"])
y_regress = m * lower_pop_df['Population density (people per sq. km of land area)'] + b

# Plot line of regression
plt.plot(lower_pop_df['Population density (people per sq. km of land area)'], y_regress, color="red", label = "line fit for all years")
plt.annotate(f"y = {m:.2f}x + {b:.2f}", color="red", xy=(65, 7.5))
plt.annotate(f"r = {r:.2f}", color="red", xy=(65, 7.0))

# create scatter plot

b = plt.scatter(lower_pop_df['Population density (people per sq. km of land area)'], lower_pop_df["Score"],c=lower_pop_df['Year'])
legend = plt.legend(*b.legend_elements(), title ='Year')
ax.add_artist(legend)
ax.legend(bbox_to_anchor=(1.1, 1.05))
plt.legend([b],labels=['line fit for all years'])

plt.title("Population Density for Lowest Happiness Scores")
plt.xlabel('Population density (people per sq. km of land area)')
plt.ylabel("Happiness Score")
plt.savefig("lowestpopulationandhappinessscore.png")
plt.show()
```
